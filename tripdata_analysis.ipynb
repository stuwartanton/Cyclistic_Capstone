```python
# Percentage of members and casual users
data['member_casual'].value_counts(normalize = True)*100
```

```python
#Percentage of bike type used for ride
data['rideable_type'].value_counts(normalize = True)*100
```

```python
#Ride Started Stations (percentages)
data['start_station_name'].value_counts(normalize = True)*100
```

```python
#Ride Ended Stations (percentages)
data['end_station_name'].value_counts(normalize = True)*100
```

```python
#Grouping by member_casual column
customer_grp = data.groupby(['member_casual'])
```

```python
# Bike type used by members and casual users (%)
customer_grp['rideable_type'].value_counts(normalize = True)*100
```

```python
# Converting ride time to minutes
data['ride_time_min'] = data['ride_time']/np.timedelta64(1,'m')
```

```python
#Ride time of Casual Users(minutes)
customer_grp['ride_time_min'].agg(['median', 'mean', 'max', 'min']).loc['casual']
```

```python
#Ride time of members(minutes)
customer_grp['ride_time_min'].agg(['median', 'mean', 'max', 'min']).loc['member']
```

```python
#Day of ride - Casual Users (%)
customer_grp['day_of_week'].value_counts(normalize = True).loc['casual']*100
```

```python
#Day of ride - Members (%)
customer_grp['day_of_week'].value_counts(normalize = True).loc['member']*100
```

```python
new_df = pd.DataFrame(data, columns = ['rideable_type', 'start_station_name', 'end_station_name','member_casual', 'day_of_week', 'ride_time_min'])
```

```python
new_df.head()
```

```python
#Data visualization
import sweetviz as sv

my_report = sv.compare_intra(new_df, new_df["member_casual"] == "casual", ["member", "casual"])

my_report.show_html()

#Data visualization is provided as html file (SWEETVIZ_REPORT.html)
```

### Observations:
- There is more members (55%) than casual users (45%)
- Classical Bike (68%) is the commonly used ride by both casual users and members
- Casual users ride longer than members
- Casual riders ride during weekends (12.5% Higher than members)

#### Suggestions for converting casual users to members
* Offers on weekend rides
* Offers on long distance rides
* Offers displayed in the primary start and end stations
