```python
# Importing all library required for analysis
import pandas as pd
import glob
import os
import numpy as np

pd.options.display.max_columns = None
```

```python
# merging the files
joined_files = os.path.join(r"C:\Users\Detective\Downloads\Google\Case_Study_1\CSV", "tripdata*.csv")
  
# A list of all joined files is returned
joined_list = glob.glob(joined_files)
  
# Finally, the files are joined
df = pd.concat(map(pd.read_csv, joined_list), ignore_index=True)
print(df)
```

```python
df = pd.read_csv('tripdata_upto_oct_2021.csv')
```

```python
# Creating a backup copy
data = df.copy()
```

```python
# Last 5 rows of the dataset
data.tail()
```

```python
# Information about the dataset
data.info()
```

```python
# Number of null values in each column
data.isnull().sum()
```

```python
# Percentage of missing values in each column
for col in data.columns:
    missing_data = np.mean(data[col].isnull()*100)
    print('{} - {}%'.format(col, round(missing_data, 2)))
```

```python
# Converting object to datatime datatype
data['started_at'] = pd.to_datetime(data['started_at'], format = '%Y-%m-%d %H:%M:%S')
data['ended_at'] = pd.to_datetime(data['ended_at'], format = '%Y-%m-%d %H:%M:%S')
```

```python
# length of the time taken for each ride
data['ride_time'] = data['ended_at'] - data['started_at']
```

```python
# Day of the week
data['day_of_week'] = data['started_at'].dt.day_name()
```

```python
# Removing negative ride time (days)
filter = (data['ride_time'] > '0')
data = data.loc[filter]
```

```python
# Cleaning Column names
data.columns = [c.strip() for c in data.columns.values.tolist()]
```

```python
# Removing any duplicate rows
data = data.drop_duplicates()
```

```python
# Drop na values
data = data.dropna()
```

```python
# Saving the cleaned dataset
data.to_csv('tripdata_upto_oct_2021_cleaned.csv')
```
